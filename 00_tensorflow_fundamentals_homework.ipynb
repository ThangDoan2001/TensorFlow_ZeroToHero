**1. Create a vector, scalar, matrix and tensor with values of your choosing using `tf.constant()`**



```python
# Create a vector
A = tf.constant(np.array([3., 7., 10.]))
A.shape[0]
```

```python
# Create a scalar
B = tf.constant(3.)
B
```

```python
# Create a matrix
C = tf.constant([[5., 9.],
                 [2., 0.],
                 [0., 1.]])
C
```

```python
# Create a tensor
D = tf.constant([[[1., 2.],
                  [3., 4.]],
                 [[5., 6.],
                  [7., 8.]]])
D
```

**2. Find the shape, rank and size of the tensors you created in 1**


```python
print("Datatype of every elements: ", D.dtype, "\n")
print("Number of dimensions(rank): ", D.ndim, "\n")
print("Shape of tensor: ", D.shape, "\n")
print("Element along the 0 axies: ", D.shape[0], "\n" )
print("Element along the last axies: ", D.shape[-1], "\n")
print("Total number in our tensor:  ", tf.size(D).numpy())
```

**3. Create two tensors containing random values between 0 and 1 with shape [5, 300]**

```python
# Create the first tensor
tf.random.set_seed(42)
E = tf.random.uniform(shape=[5, 300])
E
```

```python
# Create the second tensor
F = tf.random.uniform(shape=[5, 300], seed=42)
F
```

**4. Multiply the two tensors you created in 3 using matrix multiplication**

```python
G = tf.linalg.matmul(E, tf.reshape(F, shape=[300, 5]))
G, tf.reshape(F, shape=[300, 5])
```

```python
H = tf.linalg.matmul(E, tf.transpose(F))
H, tf.transpose(F)
```

**5. Multiply the two tensors you created in 3 using dot product**

```python
I = tf.tensordot(E, tf.reshape(F, shape=(300, 5)), axes=1)
I
```

```python
K = tf.tensordot(E, tf.transpose(F), axes=1)
K
```

**6. Create a tensor with random values between 0 and 1 with shape [224, 224, 3]**

```python
# Create a tensor
tf.random.set_seed(59)
L = tf.random.uniform(shape=(224, 224, 3))
L
```

**7. Find the min and max values of the tensor you created in 6 along the first axis**


```python
Q = tf.reduce_max(L, axis=0)
Q
```

**9. Create a tensor with shape `[10]` using your own choice of values, then find the index which has maximum value**
